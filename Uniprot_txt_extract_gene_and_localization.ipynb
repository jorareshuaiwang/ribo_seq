```python
#From parsed file, slice lines starting with ID and subcellular location:
with open('/Users/ShuaiWang/Desktop/Shan_Lab/Ribo_seq/Data_analysis/K12_parsed.txt', 'r') as f, open('/Users/ShuaiWang/Desktop/Shan_Lab/Ribo_seq/Data_analysis/K12_IDlocal.txt', 'w') as f_out:
    f.seek(0)
    f_list = f.readlines()
    for i,line in enumerate(f_list):
        if f_list[i][:6]==' ID   ' and 'ECOLI' in line[:34]:
            f_out.write(line[:19])
            f_out.write('\n')
        if line[:25]==' -!- SUBCELLULAR LOCATION':
            f_out.write(line[4:])
        
            
            

x = open('/Users/ShuaiWang/Desktop/Shan_Lab/Ribo_seq/Data_analysis/K12_IDlocal.txt', 'r')
x_list = x.readlines()
x_list[:300]
            
```
